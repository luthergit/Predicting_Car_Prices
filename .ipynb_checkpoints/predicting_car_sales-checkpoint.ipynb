# Predicting Car Sales

In this project, we will be using  machine learning workflow to predict a car's market price using its attributes. The data set we will be working with contains information on various cars. For each car we have information about the technical aspects of the vehicle such as the motor's displacement, the weight of the car, the miles per gallon, how fast the car accelerates, and more. You can read more about the data set [here](https://archive.ics.uci.edu/ml/datasets/automobile) and can download it directly from [here](https://archive.ics.uci.edu/ml/machine-learning-databases/autos/imports-85.data).

Dataset's [documentation](https://archive.ics.uci.edu/ml/datasets/automobile) lists the column names, which we will use when reading the file.

```python
import pandas as pd
import numpy as np

columns = ['symboling', 'normalized-losses', 'make', 'fuel-type', 'aspiration', 'num-of-doors', 'body-style', 
        'drive-wheels', 'engine-location', 'wheel-base', 'length', 'width', 'height', 'curb-weight', 'engine-type', 
        'num-of-cylinders', 'engine-size', 'fuel-system', 'bore', 'stroke', 'compression-rate', 'horsepower', 
           'peak-rpm', 'city-mpg', 'highway-mpg', 'price']
cars = pd.read_csv('imports-85.data', names = columns)

pd.options.display.max_columns = 99
```

```python
cars.head()
```

```python
#columns are numeric and can be used as features:
continuous_values_cols = ['normalized-losses', 'wheel-base', 'length', 'width', 'height', 'curb-weight', 'engine-size', 'bore', 'stroke', 'compression-rate', 'horsepower', 'peak-rpm', 'city-mpg', 'highway-mpg', 'price']
numeric_cars = cars[continuous_values_cols]
```

```python
numeric_cars.head()
```

## Data Cleaning

Based on the data set preview , we can tell that the `normalized-losses` column contains missing values represented using `"?"`. We will replace these values and look for the presence of missing values in other numeric columns. Also, rescale the values in the numeric columns so they all range from `0` to `1`

```python
numeric_cars = numeric_cars.replace('?', np.nan)
```

```python
numeric_cars.info()
```

```python
numeric_cars = numeric_cars.astype('float')
```

```python
numeric_cars.isnull().sum()
```

```python
numeric_cars.shape
```

```python
# Because `price` is the column we want to predict, let's remove any rows with missing `price` values.
numeric_cars = numeric_cars.dropna(subset=['price'])
numeric_cars.isnull().sum()
```

```python
numeric_cars = numeric_cars.fillna(numeric_cars.mean())
numeric_cars.isnull().sum()
```

```python
numeric_cars.head()
```

```python
# Normalize all columnns to range from 0 to 1 except the target column.
price_col = numeric_cars['price']
numeric_cars = (numeric_cars - numeric_cars.min())/(numeric_cars.max() - numeric_cars.min())
numeric_cars['price'] = price_col
```

```python
numeric_cars
```

## Univariate Model

Let's start with some univariate k-nearest neighbors models. Starting with simple models before moving to more complex models will helps structure the code workflow and understand the features better.

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error

def knn_train_test(train_col,target_col,df):
    #training
    #---------
    knn = KNeighborsRegressor()
    np.random.seed(1)
    
    #randomise row orders
    shuffled = np.random.permutation(df.index)
    rand_df = df.reindex(shuffled)
    
    #divide row in half and round up
    halves = int(len(rand_df)/2)
    
    #first half is training set
    #second half is testing set
    train_df = rand_df.iloc[: halves]
    test_df = rand_df.iloc[halves :]
    
    #fit the model on the training set
    knn.fit(train_df[[train_col]], train_df[target_col])
    
    #predicting
    #---------
    #make predictions on the test set.
    predictions = knn.predict(test_df[[train_col]])
    
    #calculate the RMSE and return that value
    rmse = np.sqrt(mean_squared_error(test_df[target_col], predictions))
    
    return rmse   
```

```python
rmse_results = {}

train_cols = numeric_cars.columns.drop('price')

#For each column (minus `price`), train a model, return RMSE value
# and add to the dictionary `rmse_results`.
for col in train_cols:
    rmse_val = knn_train_test(col,'price',numeric_cars)
    rmse_results[col] = rmse_val

# Create a Series object from the dictionary so 
# we can easily view the results, sort, etc
rmse_results_series = pd.Series(rmse_results)
rmse_results_series = rmse_results_series.sort_values()
```

Next, we'll modify the `knn_train_test()` function to accept a parameter for the k value.

```python
def knn_train_test(train_col,target_col,df):
    #training
    #---------
   
    np.random.seed(1)
    
    #randomise row orders
    shuffled = np.random.permutation(df.index)
    rand_df = df.reindex(shuffled)
    
    #divide row in half and round up
    halves = int(len(rand_df)/2)
    
    #first half is training set
    #second half is testing set
    train_df = rand_df.iloc[: halves]
    test_df = rand_df.iloc[halves :]
    
    k_value = [1, 3, 5, 7, 9]
    k_rsme = {}
    
    for k in k_value:
         #fit the model on the training set
        knn = KNeighborsRegressor(n_neighbors = k)
        knn.fit(train_df[[train_col]], train_df[target_col])
    
        #predicting
        #---------
        #make predictions on the test set.
        predictions = knn.predict(test_df[[train_col]])
    
        #calculate the RMSE and return that value
        rmse = np.sqrt(mean_squared_error(test_df[target_col], predictions))
        k_rsme[k] = rmse
    
    return k_rsme
```

```python
k_rmse_results = {}

train_cols = numeric_cars.columns.drop('price')

#For each column (minus `price`), train a model, return RMSE value
# and add to the dictionary `rmse_results`.
for col in train_cols:
    rmse_val = knn_train_test(col,'price',numeric_cars)
    k_rmse_results[col] = rmse_val

k_rmse_results
```

Visualising the results using a line plot.

```python
import matplotlib.pyplot as plt
%matplotlib inline

for k, val in k_rmse_results.items():
    x = list(val.keys())
    y = list(val.values())
    
    plt.plot(x,y)
    plt.xlabel('k_values')
    plt.ylabel('RMSE')
```

```python
# Compute average RMSE across different `k` values for each feature.

feature_rmse_avg = {}

train_cols = numeric_cars.columns.drop('price')

for k,v in k_rmse_results.items():
    avg_rmse = np.mean(list(v.values()))
    feature_rmse_avg[k] = avg_rmse

series_avg_rmse = pd.Series(feature_rmse_avg)
series_avg_rmse = series_avg_rmse.sort_values()

sorted_features = series_avg_rmse.index #we'll use this later
print(sorted_features)
series_avg_rmse
```

## Multivariate Model

Next we modify the `knn_train_test()` function to work with multiple columns.

We'll use the default n_neighors (k =5) and use the best features from (2 to 6) to train and test a multivariate k-nearest neighbors model using the default k value

```python
def knn_train_test(train_cols,target_col,df):
    #training
    #---------
   
    np.random.seed(1)
    
    #randomise row orders
    shuffled = np.random.permutation(df.index)
    rand_df = df.reindex(shuffled)
    
    #divide row in half and round up
    halves = int(len(rand_df)/2)
    
    #first half is training set
    #second half is testing set
    train_df = rand_df.iloc[: halves]
    test_df = rand_df.iloc[halves :]
    
    k_value = [5]
    k_rsme = {}
    
    for k in k_value:
         #fit the model on the training set
        knn = KNeighborsRegressor(n_neighbors = k)
        knn.fit(train_df[train_cols], train_df[target_col]) ###
    
        #predicting
        #---------
        #make predictions on the test set.
        predictions = knn.predict(test_df[train_cols])
    
        #calculate the RMSE and return that value
        rmse = np.sqrt(mean_squared_error(test_df[target_col], predictions))
        k_rsme[k] = rmse
    
    return k_rsme
```

```python
k_rmse_results = {}

for no_best_features in range(2,7):
    k_rmse_results['{} best features'.format(no_best_features)]  = knn_train_test(
        sorted_features[:no_best_features],
        'price',
        numeric_cars
    )
k_rmse_results
```

## Hyperparameter Tuning

Let's now optimize the model that performed the best.

```python
def knn_train_test(train_cols,target_col,df):
    #training
    #---------
   
    np.random.seed(1)
    
    #randomise row orders
    shuffled = np.random.permutation(df.index)
    rand_df = df.reindex(shuffled)
    
    #divide row in half and round up
    halves = int(len(rand_df)/2)
    
    #first half is training set
    #second half is testing set
    train_df = rand_df.iloc[: halves]
    test_df = rand_df.iloc[halves :]
    
    k_value = [i for i in range(1,26)]
    k_rsme = {}
    
    for k in k_value:
         #fit the model on the training set
        knn = KNeighborsRegressor(n_neighbors = k)
        knn.fit(train_df[train_cols], train_df[target_col])
    
        #predicting
        #---------
        #make predictions on the test set.
        predictions = knn.predict(test_df[train_cols])
    
        #calculate the RMSE and return that value
        rmse = np.sqrt(mean_squared_error(test_df[target_col], predictions))
        k_rsme[k] = rmse
    
    return k_rsme
```

```python
k_rmse_results = {}

for no_best_features in range(2,6):
    k_rmse_results['{} best features'.format(no_best_features)]  = knn_train_test(
        sorted_features[:no_best_features],
        'price',
        numeric_cars
    )
k_rmse_results
```

```python
for k, val in k_rmse_results.items():
    x = list(val.keys())
    y = list(val.values())
    plt.plot(x,y,label = '{}'.format(k))
    
    plt.xlabel('k_values')
    plt.ylabel('RMSE')
    plt.legend()
```

* `k` value of 2 is the optimal model for the two best features
* `k` value of 3 is the optimal model for the three best features
* `k` value of 1 is the optimal model for the four best features
* `k` value of 1 is the optimal model for the fifth best features

The number of best features included, affects the k value. This means that selecting the right features is important and that using more features doesn't automatically improve prediction accuracy
